```python
class Score:
  def __init__(self, value):
    self.value = value
  def __eq__(self, other):
    return self.value == other.value
  def __lt__(self, other):
    return self.value < other.value
  def __add__(self, other):
    return Score(self.value + other.value)

a = Score(10)
b = Score(20)

print(a == b)
print(a < b)
c = a + b
c.value
```

Part 1.1: Closures

```python
def outer():
  x = 10
  def inner():
    print("x =", x)
  return inner

f = outer()
f()
f()
```

Part 1.2: Closure with Changing State

```python
def counter():
  count = 0
  def inc():
    nonlocal count
    count += 1
    print("count =", count)
  return inc

c = counter()
c()
c()
c()
```

Part 1.3: Prove Where the Data Lives

```python
print(c.__closure__)
print(c.__closure__[0].cell_contents)
```

Decorators

Part 2.1: Basic Decorator (Wrapper in Action)

```python
def my_decorator(func):
  def wrapper():
    print("Before Function")
    func()
    print("After Function")
  return wrapper

@my_decorator
def greet():
  print("Hello")

greet()
```

Part 2.2: Decorator vs Wrapper (IMPORTANT)

```python
def demo_decorator(func):
  print("DECORATOR runs (defenition time)")
  def wrapper():
    print("WRAPPER start")
    func()
    print("WRAPPER end")
  return wrapper

@demo_decorator
def say_hi():
  print("HI")
```

```python
say_hi()
```

Part 2.3: Inspect the Decorated Function

```python
print(say_hi)
print(say_hi.__closure__)
```

Decorator with Arguments

Part 3: Decorator with Arguments

```python
def log_calls(func):
  def wrapper(*args):
    print("Calling with args:", args)
    return func(*args)
  return wrapper

@log_calls
def add(a, b):
  return a + b

print(add(3, 4))
print(add(10,20))
```

Closure vs Decorator

Part 4.1: Closure = Data Memory

```python
def make_multiplier(n):
  def multiply(x):
    return x * n
  return multiply

double = make_multiplier(2)
triple = make_multiplier(3)

print(double(5))
print(triple(5))
```

Part 4.2: Decorator = Behavior Wrapping

```python
def loud(func):
  def wrapper():
    print("LOUD MODE")
    func()
  return wrapper

@loud
def speak():
  print("hello")

speak()
```

Final Rule: Closures store data. Decorators configure behavior. Wrappers execute behavior.
